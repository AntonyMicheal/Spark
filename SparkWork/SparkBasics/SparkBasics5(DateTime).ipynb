# ***Lets find average close per year***

```python
df.select(year(df['date'])).show()
```

```python
df.withColumn("Year", year(df['date'])).show()
```

```python
new_df = df.withColumn("Year", year(df['date']))
```

```python
new_df.groupBy("Year").mean().show()
```

```python
result = new_df.groupBy("Year").mean().select(["Year", "avg(close)"]).sort('year')
```

```python
result.show(26)
```

```python
new = result.withColumnRenamed("avg(close)", "Average closing price")
```

```python
new.show()
```

```python
new.select(['year',format_number('Average closing price', 2).alias('Average Close')]).show()
```
